```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/faisal-apew/Pembelajaran-Mesin/master/ipl_schedule.csv'
df = pd.read_csv(url)
```

```python
df.sample(4)
```

```python
df.info()
```

```python
df['DATE']=df['DATE'].astype("datetime64[ns]")
```

```python
Time_frequecy=pd.DataFrame(df.TIME.value_counts(normalize=True)*100).reset_index()
Time_frequecy.plot.bar("index","TIME",title="IPL 2022 Match Timings.")
```

```python
df_time=pd.DataFrame(df['DATE'].value_counts()).reset_index()
df_time.DATE.value_counts()
```

```python
Home_team=pd.DataFrame(df['HOME TEAM'].value_counts()).reset_index()
Home_team.rename(columns={'index':'Team Name'},inplace=True)
Home_team
```

```python
AWAY_team=pd.DataFrame(df['AWAY TEAM'].value_counts()).reset_index()
AWAY_team.rename(columns={'index':'Team Name'},inplace=True)
AWAY_team
```

```python
VENUE=pd.DataFrame(df['VENUE'].value_counts()).reset_index()
VENUE.rename(columns={'index':'Venue','VENUE':'Matches to be Played'},inplace=True)
VENUE
```
